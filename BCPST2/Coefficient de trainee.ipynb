# 1 - Coefficient de traînée
---

Ce document aborde l'influence du nombre de Reynolds sur la valeur du coefficient de traînée **$\mathsf{C_x}$** lors de la chute d'une sphère dans un fluide newtonien. Le coefficient de traînée intervient dans l'expression de la norme de la force de frottement quadratique :

\begin{equation}   
   \mathsf{ 
   \left\|\vec{F}\right\| = \frac{1}{2} C_x \rho v^2 \pi r^2
   }
\end{equation}

Notations : 
* $\mathsf{\rho}$ : masse volumique du fluide dans lequel s'effectue la chute,
* $\mathsf{v}$ : vitesse de chute de l'objet sphérique,
* $\mathsf{r}$ : rayon de la sphère.

Des mesures en aérodynamique permettent d'adapter la valeur du coefficient de traînée $\mathsf{C_x}$ au nombre de Reynolds diamétral :

\begin{equation}   
   \mathsf{ 
   R_e = \frac{2 \rho v r}{\eta}
   }
\end{equation}

avec $\mathsf{\eta}$ : viscosité dynamique du fluide


Plusieurs cas sont envisagés :

\begin{array}{l|c|c}
    \mathsf{\text{Nature de l'}\acute{e}{coulement}} & \rm \mathsf{\text{Condition}} & \rm \mathsf{\text{Expression de }R_e} \\ \hline
    \mathsf{\text{Ecoulement rampants}} & \mathsf{R_e \leqslant 1} & \mathsf{C_x = \frac{24}{Re} } \\ \hline
    \mathsf{\text{Ecoulement intermédiaire}} & \mathsf{1< R_e \leqslant 1 \cdot 10^3} & \mathsf{C_x = \frac{18.5}{R_e^{0.6}} } \\ \hline
    \mathsf{\text{Plage sous-critique}} & \mathsf{1 \cdot 10^3 < Re \leqslant 3 \cdot 10^5} & \mathsf{C_x \approx 0.5} \\ \hline
    \mathsf{\text{Plage super critique}} & \mathsf{R_e > 3 \cdot 10^5} & \mathsf{C_x \approx 0.07} \\ 
\end{array}

Dans le cas des écoulements rampants, la force de traînée est bien proportionnelle à la vitesse :

\begin{equation}     
   \begin{cases} 
       \mathsf{\left\| \vec{F} \right\| = \frac{1}{2} C_x \rho v^2 \pi r^2} \\ 
       \mathsf{C_x = \frac{24}{Re}} \\
       \mathsf{R_e = \frac{2 \rho v r}{\eta}} \\
   \end{cases}
   \Rightarrow
   \mathsf{ \left\| \vec{F} \right\| = 6 \pi \eta r v}
\end{equation}

# 2 - Equation du mouvement dans le cas d'une chute verticale
---

Dans le référentiel terrestre, supposé galiléen pendant la durée de l'expérience, l'objet sphérique, de masse volumique $\mathsf{\mu}$ est soumis, dans un fluide de masse volumique $\mathsf{\rho}$ à :
- son poids $\vec{P}$,
- la poussée d'Archimède $\vec{\Pi}$,
- une force de frottement fluide $\vec{F}$.

L'application de la seconde loi de Newton, puis sa projection selon un axe z orienté vers le bas, conduit à l'équation du mouvement suivante :

\begin{equation}   
   \mathsf{ 
   m \cdot \ddot{z} = m \cdot \left( 1-\frac{\rho}{\mu} \right) \cdot g - \frac{1}{2} C_x \rho v^2 \pi r^2 \\
   \ddot{z} = \left( 1 - \frac{\rho}{\mu} \right) \cdot g - \frac{1}{2m} C_x \rho v^2 \pi r^2 \\
   }
\end{equation}
où $\mathsf{v^2=\left( \dot{z} \right)^2}$.


L'origine de l'axe des $\mathsf{z}$ est prise à l'altitude initiale. A un instant $\mathsf{t}$ quelconque, l'altitude vaut donc $\mathsf{h(t)=h_0-z}$.

# 3 - Résolution au moyen de la méthode d'Euler
---

Dans la méthode d'Euler, les valeurs successives de la vitesse sont obtenues grâce à la formule de Taylor :

\begin{equation}
    \mathsf{
    {\dot z}_{t+a}= {\dot z}_{t} + \left( \ddot{z} \right)_{t}  \cdot a 
    }
\end{equation}

L'altitude de la bille s'obtient ensuite par une intégration analogue : 

\begin{equation}
    \mathsf{
    z_{t+a} = z_{t} + \left( \frac{dz}{dt} \right)_{t}  \cdot a 
    \text{         } \Rightarrow \text{         }
    h_{t+a} = h_{t} - \left( \frac{dz}{dt} \right)_{t}  \cdot a
    }
\end{equation}



### Manipulation du script

* Caractéristiques de la bille ($\mathsf{\mu}$ et $\mathsf{r}$) ou du fluide ($\mathsf{\rho}$ et $\mathsf{\eta}$),
* Conditions initiales ($\mathsf{h_0}$ et $\mathsf{zpoint_0}$)
* Pas de calcul $\mathsf{a}$ de la méthode d'Euler.


## Etude d'un cas

Les paramètres numériques choisis sont les suivants :
* **Données relatives à l'objet sphérique (bille en acier)** 
    * Masse volumique : $\mathsf{ \mu=7.8 \cdot 10^3 kg \cdot m^{-3} }$,
    * Rayon : $\mathsf{ r=6 \cdot 10^{-3} m}$,
* **Données relatives au fluide (glycérol)**
    * Masse volumique : $\mathsf{ \rho=1.3 \cdot 10^3 kg \cdot m^{-3} }$,
    * Viscosité dynamique : $\mathsf{ \eta=1.5 Pl}$,
* **Conditions initiales**
    * Altitude initiale : $\mathsf{h_0}$ 
    * Vitesse initiale : $\mathsf{zpoint_0}$ (positive si la bille est envoyée vers le bas, négative si elle est envoyée vers le haut).

```python
#Importation des bibliothèques
import matplotlib.pyplot as plt
from math import pi


#Données numériques
g=9.8       #accélération de la pesanteur en m/s2
mu=7.8e3    #masse volumique de la bille en kg/m3
r=6e-3      #rayon de la bille en m
rho=1.2e3   #masse volumique du fluide en kg/m3
eta=1.5     #viscosité dynamique du fluide en Pl
z0=0.5      #altitude initiale (en m)
zpoint0=.000001  #composante verticale initiale de la bille en m/s

#Calcul de la masse de la bille
m = 4/3 * mu * pi * r**3

#Calcul de section droite de la bille
A = pi * r**2

#Paramètres de la méthode d'intégration
deltat=0.001     # pas de calcul en s
duree = .3       # durée d'intégration en s


#Initialisation de listes
LT=[0]                                            #Liste des dates successives
Lzpoint=[zpoint0]                                 #Liste des composantes verticales successives de la vitesse
Lv = [abs(zpoint0)]                               #Liste des vitesses successives
Lz=[z0]                                           #Liste des positions successives
LRe=[2*rho*abs(zpoint0)*r / eta]                  #Liste des nombres de Reynolds successifs


#Définition d'une fonction permettant de calculer l'intensité de la force de frottements
def F(Re) :
    if Re<=.1 :
        Cx = 24/Re
    elif .1<Re<=1000 :
        Cx = 24/Re*(1+.14*Re**.7)
    elif 1000<Re<=3e5 :
        Cx = .445
    else :
        Cx = .19 - 8e4/Re
    return -.5 * Cx * A * rho * zpoint * abs(zpoint)


#Boucle d'intégration Euler
nbpts=duree/deltat
i=1
while i<=nbpts :
    zpoint=Lzpoint[-1]
    z=Lz[-1]
    
    Re = 2*r * rho * abs(zpoint) / eta
    
    LT.append(i * deltat)
    Lzpoint.append(zpoint + deltat * (-(1-rho/mu)*g + F(Re)/m))
    Lz.append(z + zpoint * deltat)
    Lv.append(abs(Lzpoint[-1]))
    LRe.append(Re)
    
    i=i+1

    
#Tracé des courbes simulées : vitesse et altitude en fonction de temps

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figheight(5)
fig.set_figwidth(18)

ax1.plot(LT,Lz,'b-', label='Position')
ax1.set_xlabel('temps (s)')
ax1.set_ylabel('z (m)')
ax1.legend()
ax1.grid(True)

ax2.plot(LT,Lv,'g-', label='Vitesse')
ax2.set_xlabel('temps (s)')
ax2.set_ylabel('vitesse en m.s$^{-1}$')
ax2.legend()
ax2.grid(True)

ax3.plot(LT,LRe,'r-', label='Nombre de Reynolds')              
ax3.set_xlabel('temps (s)')
ax3.set_ylabel('Re')
ax3.legend()
ax3.grid(True)

plt.show()
```

## Influence de la vitesse initiale (avec curseur)

Ce script, analogue au précédent, illustre le fait que quelle que soit la vitesse initiale communiquée au projectile, celle-ci tend à adopter une valeur constante au bout d'une certaine durée.
Un bouton interactif permet de modifier la valeur de la vitesse initiale.

```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
from ipywidgets import interact


#Données
g=9.8       #accélération de la pesanteur en m/s2
mu=7.8e3    #masse volumique de la bille en kg/m3
r=1e-2      #rayon de la bille en m
z0 = 0.5    #altitude initiale en m
rho=1.2e3   #masse volumique du fluide en kg/m3
eta=1.5     #viscosité dynamique du fluide en Pl

#Calcul de la masse de la bille
m = 4/3 * mu * pi * r**3

#Calcul de section de la bille
A = pi * r**2


#Définition d'une fonction permettant de calculer l'intensité de la force de frottements
def F(Re,rho,eta,zpoint) :
    if Re<=.1 :
        Cx = 24/Re
    elif .1<Re<=1000 :
        Cx = 24/Re*(1+.14*Re**.7)
    elif 1000<Re<=3e5 :
        Cx = .445
    else :
        Cx = .19 - 8e4/Re
    return -.5 * Cx * A * rho * zpoint * abs(zpoint)

      
    
#Procédure d'intégration et de tracé
def ChuteVitesse(zpoint0):
    deltat=0.0001
    nbpts=3000
    
    #Initialisation de listes
    zpoint0_modif = zpoint0 + 0.0001
    #L'ajout de 0.00001 m/s permet de ne pas diviser par zéro lors du calcul de Cx
    LT=[0]                                      #Liste des dates successives
    Lzpoint=[zpoint0_modif]                     #Liste des composantes verticales successives de la vitesse

    #Boucle d'intégration
    i=1
    while i<=nbpts :
        zpoint=Lzpoint[-1]
        
        Re = 2*r * rho * abs(zpoint) / eta

        LT.append(i * deltat)
        Lzpoint.append(zpoint + deltat * (-(1-rho/mu)*g + F(Re,rho,eta,zpoint)/m))

        i=i+1

    #Création d'une figure
    plt.figure()
    plt.plot(LT,Lzpoint,'.')
    plt.xlabel('temps (s)')
    plt.ylabel('$\dot{z}$ (m/s)')
    plt.ylim(-2.1,1.1)
    plt.axhline(-0.58,ls='--',color='r')
    plt.title('Vitesse en fonction du temps')
    plt.grid(True)

interact(ChuteVitesse,zpoint0=([1, 0,-2]))
```

## Influence de la nature du fluide (avec curseur)

Ce script illustre l'influence de la nature du fluide sur le nombre de Reynolds et sur le modèle privilégié de la force de frottement fluide.

```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
from ipywidgets import interact


#Données
g=9.8       #accélération de la pesanteur en m/s2
mu=7.8e3    #masse volumique de la bille en kg/m3
r=1e-3      #rayon de la bille en m
z0 = 0.5    #altitude initiale en m

#Calcul de la masse de la bille
m = 4/3 * mu * pi * r**3

#Calcul de section de la bille
A = pi * r**2


#Définition d'une fonction permettant de calculer l'intensité de la force de frottements
def F(Re,rho,eta,zpoint) :
    if Re<=.1 :
        Cx = 24/Re
    elif .1<Re<=1000 :
        Cx = 24/Re*(1+.14*Re**.7)
    elif 1000<Re<=3e5 :
        Cx = .445
    else :
        Cx = .19 - 8e4/Re
    return -.5 * Cx * A * rho * zpoint * abs(zpoint)


#Définition d'une fonction permettant de retourner les valeurs caractéristiques du fluide
def Caract(fluide) :
    if fluide == 'eau' :
        rho=1e3
        eta=1e-3
        duree=.3
    elif fluide == 'air' :
        rho=1
        eta=1.5e-5
        duree=.3
    elif fluide == 'glycerol' :
        rho=1.2e3
        eta=1.5
        duree=.02
    return rho,eta,duree
        
    
#Procédure d'intégration et de tracé
def ChuteFluide(fluide):
    
    rho,eta,duree=  Caract(fluide)
    nbpts=5000
    deltat=duree/nbpts
    
    #Initialisation de listes
    LT=[0]                                        #Liste des dates successives
    Lzpoint=[0.00001]                             #Liste des composantes verticales successives de la vitesse
    Lz=[z0]                                       #Liste des altitudes successives
    Lv=[0.00001]                                  #Liste des vitesses successives
    LRe=[2*r * rho * abs(zpoint0) / eta]          #Liste des nombres de Reynolds successifs
    LF=[0]                                        #Liste des normes successives de la force de frottement

    #Boucle d'intégration
    i=1
    while i<=nbpts :
        zpoint=Lzpoint[-1]
        z=Lz[-1]
        
        Re = 2*r * rho * abs(zpoint) / eta

        LT.append(i * deltat)
        Lzpoint.append(zpoint + deltat * (-(1-rho/mu)*g + F(Re,rho,eta,zpoint)/m))
        Lz.append(z + zpoint * deltat)
        LRe.append(Re)
        Lv.append(abs(Lzpoint[-1]))
        LF.append(F(Re,rho,eta,zpoint))
        i=i+1

    #Création d'une figure
    fig, axs = plt.subplots(2,2)
    fig.set_figheight(12)
    fig.set_figwidth(13)

    #Commandes de tracé
    axs[0,0].plot(LT,Lz)
    axs[0,0].set_xlabel('temps (s)')
    axs[0,0].set_ylabel('z(m)')
    axs[0,0].set_title('Altitude en fonction du temps')
    axs[0,0].grid(True)
                  
    axs[0,1].plot(LT,Lv)
    axs[0,1].set_ylabel('Vitesse en m.s$^{-1}$')
    axs[0,1].set_xlabel('temps (s)')
    axs[0,1].set_title('Vitesse en fonction du temps')
    axs[0,1].grid(True)

    axs[1,0].plot(LT,LRe) 
    axs[1,0].set_ylabel('Re')
    axs[1,0].set_xlabel('temps (s)')
    axs[1,0].set_title('Nombre de Reynolds en fonction du temps')
    axs[1,0].grid(True)
    
    axs[1,1].plot(Lv,LF)
    axs[1,1].set_xlabel('vitesse (m/s)')
    axs[1,1].set_ylabel('Force (N)')
    axs[1,1].set_title('Force en fonction de la vitesse')
    axs[1,1].grid(True)

interact(ChuteFluide,fluide=(['air', 'eau','glycerol']))
```

## Influence du rayon de la bille (avec curseur)

Ce script, analogue aux précédents, illustre l'influence du rayon de la bille sur le nombre de Reynolds et par conséquent, sur le modèle privilégié de la force de frottement fluide.


```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
from ipywidgets import interact


#Données
g=9.8       #accélération de la pesanteur en m/s2
mu=7.8e3    #masse volumique de la bille en kg/m3
z0 = 0.5    #altitude initiale en m
rho=1.2e3   #masse volumique du fluide en kg/m3
eta=1.5     #viscosité dynamique du fluide en Pl
duree=.1

#Calcul de la masse de la bille
m = 4/3 * mu * pi * r**3

#Calcul de section de la bille
A = pi * r**2

#Définition d'une fonction permettant de calculer l'intensité de la force de frottements
def F(Re,rho,eta,zpoint) :
    if Re<=.1 :
        Cx = 24/Re
    elif .1<Re<=1000 :
        Cx = 24/Re*(1+.14*Re**.7)
    elif 1000<Re<=3e5 :
        Cx = .445
    else :
        Cx = .19 - 8e4/Re
    return -.5 * Cx * A * rho * zpoint * abs(zpoint)
        
    
#Procédure d'intégration et de tracé
def ChuteRayon(rayon):
    r=rayon
    nbpts=10000
    deltat=duree/nbpts
    
    #Initialisation de listes
    LT=[0]                                        #Liste des dates successives
    Lzpoint=[0.00001]                             #Liste des composantes verticales successives de la vitesse
    Lv=[0.00001]                                  #Liste des vitesses successives
    LRe=[2*r * rho * abs(zpoint0) / eta]          #Liste des nombres de Reynolds successifs
    LF=[0]                                        #Liste des normes successives de la force de frottement

    #Boucle d'intégration
    i=1
    while i<=nbpts :

        zpoint=Lzpoint[-1]
        Re = 2*r * rho * abs(zpoint) / eta

        LT.append(i * deltat)
        Lzpoint.append(zpoint + deltat * (-(1-rho/mu)*g + F(Re,rho,eta,zpoint)/m))
        LRe.append(Re)
        LF.append(F(Re,rho,eta,zpoint))
        Lv.append(abs(Lzpoint[-1]))

        i=i+1

    #Création d'une figure
    fig, (ax1,ax2) = plt.subplots(1,2)
    fig.set_figheight(5)
    fig.set_figwidth(13)

    #Commandes de tracé
    ax1.plot(Lv,LF)
    ax1.set_xlabel('vitesse (m/s)')
    ax1.set_ylabel('Force (N)')
    ax1.set_title('Force en fonction de la vitesse')
    ax1.grid(True)
                  
    ax2.plot(LT,LRe)
    ax2.set_ylabel('Nombre de Reynolds')
    ax2.set_xlabel('temps (s)')
    ax2.set_title('Nombre de Reynolds en fonction du temps')
    ax2.grid(True)

interact(ChuteRayon,rayon=([2e-3,2e-2,2e-1]))

```

# 4 - Intégration au moyen de la fonction `odeint`
---

Le script est écrit pour illustrer l'influence de la nature du fluide sur le mouvement de la bille et le modèle de frottements fluides le plus adapté.

```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy.integrate import odeint


#Données
g=9.8       #accélération de la pesanteur en m/s2
mu=7.8e3    #masse volumique de la bille en kg/m3
r=6e-3      #rayon de la bille en m
z0=.5       #altitude initiale de la bille en m
zpoint0=.00001   #composante vertiacale initiale de la bille en m/s


#Calcul de la masse de la bille
m = 4/3 * mu * pi * r**3

#Calcul de section de la bille
A = pi * r**2


#Définition d'une fonction permettant de calculer l'intensité de la force de frottements
def F(rho,eta,zpoint) :
    Re=2*r*abs(zpoint)*rho/eta
    if Re<=.1 :
        Cx = 24/Re
    elif .1<Re<=1000 :
        Cx = 24/Re*(1+.14*Re**.7)
    elif 1000<Re<=3e5 :
        Cx = .445
    else :
        Cx = .19 - 8e4/Re
    return -.5 * Cx * A * rho * zpoint * abs(zpoint)


#Définition d'une fonction permettant de retourner les valeurs caractéristiques du fluide"""
def Caract(fluide) :
    if fluide == 'eau' :
        rho=1e3
        eta=1e-3
        duree=.4
    elif fluide == 'air' :
        rho=1
        eta=1.5e-5
        duree=.3
    elif fluide == 'glycerol' :
        rho=1.2e3
        eta=1.5
        duree=.2
    return rho,eta,duree


    
""" Procédure d'intégration et de tracé"""
def ChuteOdeint(fluide):
    
    rho,eta,duree =  Caract(fluide)

    #Création d'une base de temps
    t=np.linspace(0,duree,4000)
     
    def f(X,t) :
        z,zpoint = X
        return zpoint, -(1-rho/mu) * g + 1/m*F(rho,eta,zpoint)
    
    result=odeint(f,[z0,zpoint0],t)
    
    Lz = result[:,0]
    Lv = abs(result[:,1])
    LRe = 2*r*Lv*rho/eta

    #Création d'une figure
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.set_figheight(5)
    fig.set_figwidth(18)

    #Commande de tracé    
    ax1.plot(t,Lz,'b-', label='Altitude')
    ax1.set_xlabel('temps (s)')
    ax1.set_ylabel('z (m)')
    ax1.legend()
    ax1.grid(True)
                  
    ax2.plot(t,Lv,'g-', label='Vitesse')
    ax2.set_xlabel('temps (s)')
    ax2.set_ylabel('Vitesse en m.s$^{-1}$')
    ax2.legend()
    ax2.grid(True)

    ax3.plot(t,LRe,'r-', label='Nombre de Reynolds')              
    ax3.set_xlabel('temps (s)')
    ax3.set_ylabel('Re')
    ax3.legend()
    ax3.grid(True)

interact(ChuteOdeint,fluide=(['air', 'eau','glycerol']))
```

# 5 - Conclusion
---

La prise en compte des frottements fluides permet d'illustrer les notions de régime transitoire et de régime établi. Au bout d'une certaine durée, la bille en acier atteint une vitesse limite de chute. Passé le régime transitoire, l'altitude décroît de manière affine, ce qui est caractéristique d'un mouvement vertical uniforme.
